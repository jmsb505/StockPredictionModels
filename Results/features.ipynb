```python
df=pd.read_csv(project_path+'/Data/Complete.csv')
df_dict={}
for key in df['ticker_symbol'].unique():
    df_dict[key]=df[df['ticker_symbol']==key]
    df_dict[key]=df_dict[key].drop(columns=['ticker_symbol'])
    df_dict[key]=df_dict[key].sort_values(by=['Date']).reset_index(drop=True)
    #df_dict[key]=df_dict[key].drop(columns=['Date'])
keys=df_dict.keys()
keys_list=list(keys)
for ticker_symbol in keys_list:
# %%
    ticker=ticker_symbol
    print(ticker)
    df=df_dict[ticker].copy()
    df=df[['Date', 'p_sentiment', 'Open', 'High', 'Low',
    'Volume', 'unrate', 'psr', 'm2', 'dspic', 'pce', 'reer', 'ir', 'ffer',
    'tcs', 'indpro', 'ccpi', 'Close']]

    dates = pd.to_datetime(df['Date'])

    # %%
    cols=list(df)[1:]


    # %%
    df_for_training = df[cols].astype(float)

    # %%
    scaler = MinMaxScaler(feature_range=(0, 1))
    df_normalized = pd.DataFrame(scaler.fit_transform(df_for_training), columns=df_for_training.columns)
    print(ticker)
    threshold=16
    high_score_features = []
    df_normalized_features=df_normalized[['p_sentiment', 'Open', 'High', 'Low',
    'Volume', 'unrate', 'psr', 'm2', 'dspic', 'pce', 'reer', 'ir', 'ffer',
    'tcs', 'indpro', 'ccpi']]
    df_normalized_target=df_normalized[['Close']]
    feature_scores=fs.mutual_info_regression(df_normalized_features,df_normalized_target, random_state=0)
    for score, f_name in sorted(zip(feature_scores, df_normalized_features.columns), reverse=True)[:threshold]:
        print(f_name, score)
        high_score_features.append(f_name)

    

```
